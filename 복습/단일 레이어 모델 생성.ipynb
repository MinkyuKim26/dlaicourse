```python
# 딥러닝 모델 생성 연습
# 딥러닝 모델을 만들기 위해선 입력값과 출력값이 필요하다. 그리고 이를 학습할 모델이 있어야지.
# 학습시킬 모델은 레이어를 갖고 있다. 내가 가지고 있는 데이터 종류에 따라 적합한 레이어가 다르다
# 모델의 레이어를 쌓은 뒤에는 최적화 방식과 손실 함수가 필요하다. 최적화 방식과 손실 함수는 모델의 레이어들이 지니고 있는 가중치(혹은 셀)을 학습시키는데 사용된다. 이 때 학습되는 비율을 최적화 방식에 적용시키면 학습되는 정도를 다르게 설정할 수 있다. 
# 최적화 함수와 손실 함수를 지정한 다음 metrics = ['acc']을 넣어주면 결과값을 그래프로 출력할 때 편하다. 
# 다음으로 훈련을 시랭한다. 훈련을 할 때 전달해야할 값은 (입력값, 출력값, 훈련 횟수, 검증 셋)이다. 훈련을 마친 뒤 반환되는 값은 훈련 기록이며 이를 history라는 변수에 저장하는 방식이 일반적이다.
# 마지막으로 history 객체와 matplotlib을 통해 그래프를 출력한다. 그래프를 보며 훈련 정도를 확인할 수 있다. 
```

```python
# 모델 생성
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
model = tf.keras.Sequential() # 모델 초기화
```

```python
model.add(keras.layers.Dense(1, input_shape = [1]) ) # 레이어 추가할 때 필요한거 : 어떤 레이어인가? 유닛(셀)은 몇개 있나? (첫번 째 레이어일 경우) 입력값의 모양은 무엇인가? 
# 그저 값만 구하는거면 할성화 함수가 필요없다. 0~1사이 값을 얻을 이유가 없으니. 근데 뭔가 판단하려면 활성화 함수가 필요하다. 
# 만약 다중 분류라면 softmax를, 이진 분류라면 sigmoid를 쓴다. 마지막 레이어인 경우 유닛의 개수 == 분류할 클래스의 개수이니 주의하자
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error', metrics = ['acc']) # 컴파일 함수에서 최적화 방식(학습 방식)이랑 손실 함수를 정한다. 
```

```python
# 데잉터 준비
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float) # 훈련시키려면 넘파이 배열이 필요하다. 아니면 Image_Generator로 만든 객체가 필요하다. 
```

```python
model.fit(Xs, Ys, epochs=500)
```

```python
model.predict([10.0]) # 값 예측. 가장 많이 쓰일 함수가 아닐까. 역시 입력값으로 넘파이 배열을 요구한다. 
```
